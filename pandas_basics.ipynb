# Python Pandas Basics


**Pandas** is a powerful library for data analysis and manipulation in Python.

It provides two main data structures:
- **Series**: A one-dimensional array-like object.
- **DataFrame**: A two-dimensional table with labeled axes (rows and columns).
        

```python

# Importing pandas
import pandas as pd
```

## Creating a Series

```python

# Creating a Series from a list
data = [10, 20, 30, 40, 50]
series = pd.Series(data)
series
```

## Creating a DataFrame

```python

# Creating a DataFrame from a dictionary
data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35],
    "City": ["New York", "Los Angeles", "Chicago"]
}
df = pd.DataFrame(data)
df
```

## Exploring Data

```python

# Display the first few rows
df.head()

# Display the shape of the DataFrame
print("Shape:", df.shape)

# Display summary statistics
df.describe()
```

## Selecting Data

```python
# Selecting a single column
df["Name"]
```

```python
# Selecting multiple columns
df[["Name", "City"]]
```

```python
# Selecting rows by index
df.iloc[0]
```

## Filtering Data

```python
# Filtering rows where Age is greater than 25
filtered_df = df[df["Age"] > 25]
filtered_df
```

## Adding a New Column

```python

# Adding a new column
df["Salary"] = [50000, 60000, 70000]
df
```
